```python
# example of brighting image augmentation
from tqdm.notebook import tqdm
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
import cv2
import os
import glob
import numpy as np

for dir_num in range(110):
    # 이미지 로드
    img = load_img('/content/drive/MyDrive/표지판/dataset/{}.jpg'.format(dir_num))
    # Numpy array 로 변환
    data = img_to_array(img)
    # expand dimension to one sample
    samples = expand_dims(data, 0)
    # image data augmentation generator 생성
    datagen = ImageDataGenerator(
        brightness_range=[0.2, 1.4],
        zoom_range=[0.6, 1])
    # prepare iterator
    it = datagen.flow(samples, batch_size=1)
    for i in range(30):
            batch = it.next()
            image = batch[0].astype("uint8")
            # rgb 변환
            b, g, r = cv2.split(image)
            img_astro3_rgb = cv2.merge([r, g, b])
            cv2.imwrite("/content/drive/MyDrive/표지판/dataset/traffic_sign/{}_{}.png".format(dir_num, i), img_astro3_rgb)
```

```python
from PIL import Image
import os
import glob
import numpy as np
from sklearn.model_selection import train_test_split

path = '/content/drive/MyDrive/표지판/dataset/traffic_image/'
full_names = os.listdir(path)
labels = sorted([each.split('.')[0] for each in full_names])
caltech_dir = '/content/drive/MyDrive/표지판/dataset/traffic_image/'
categories = labels
nb_classes = len(labels)

image_w = 64
image_h = 64

X = []
y = []

for idx, cat in enumerate(categories):
    #   one-hot 돌리기.
    label = [0 for i in range(nb_classes)]
    label[idx] = 1
    image_dir = caltech_dir + "/" + str(cat)
    files = glob.glob(image_dir+"/*.png")
    print(cat, " 파일 길이 : ", len(files))
    # 이미지 파일을 64 x 64 로 줄이고, 벡터화 시켜 X에 저장, one-hot-encoding된 라벨도 저장
    for i, f in enumerate(files):
        img = Image.open(f)
        img = img.convert("RGB")
        img = img.resize((image_w, image_h))
        data = np.asarray(img)

        X.append(data)
        y.append(label)

X = np.array(X)
y = np.array(y)
        
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
xy = (X_train, X_test, y_train, y_test)
```

```python
X_train, X_test, y_train, y_test = np.load('/content/drive/MyDrive/표지판/dataset/multi_image_data.npy',allow_pickle=True)
```

```python
# 픽셀 값 정규화
X_train = X_train.astype(float) / 255
X_test = X_test.astype(float) / 255
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import layers
model = Sequential()

model.add(layers.InputLayer(input_shape=(64,64,3)))
model.add(layers.Conv2D(16,(3,3),(1,1),'same',activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(rate=0.3))

model.add(layers.Conv2D(32, (3, 3), (1, 1), 'same', activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(rate=0.3))

model.add(layers.Conv2D(64, (3, 3), (1, 1), 'same', activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(rate=0.3))

model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(2, activation='sigmoid'))

model.summary()

# 학습을 돌리는 방법을 정의 : cost function을 설정하고, 어떻게 최적화 할건지 방법을 정하고
model.compile(loss='categorical_crossentropy',
              optimizer='adam', metrics=['accuracy'])
model_dir = '/content/drive/MyDrive/표지판/model2'

if not os.path.exists(model_dir):
      os.mkdir(model_dir)

model_path = model_dir + '/content/drive/MyDrive/표지판/model2/multi_img_classification.model'
checkpoint = ModelCheckpoint(
    filepath=model_path, monitor='val_loss', verbose=1, save_best_only=True)
early_stopping = EarlyStopping(monitor='val_loss', patience=6)
```

```python
history = model.fit(X_train, y_train, batch_size=32, epochs=10)
```

```python
import matplotlib.pyplot as plt
plot_target = ['loss', 'val_loss', 'acc', 'val_acc']
for each in plot_target:
    plt.plot(history.history[each], label=each)
    
plt.legend()
plt.show()
```

```python
model.evaluate(X_test, y_test)
```
